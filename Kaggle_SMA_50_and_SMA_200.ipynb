## BACK TESTING

```python
def calculate_profit(investment, initial_price, future_price):
    initial_value = investment / initial_price
    future_value = initial_value * future_price
    profit = future_value - investment
    return profit
```

```python
def plot_profit_over_time(dates, profits):
    plt.figure(figsize=(15, 6))
    above_zero_profits = [profit if profit >= 0 else 0 for profit in profits]
    below_zero_profits = [profit if profit < 0 else 0 for profit in profits]

    plt.stackplot(dates, above_zero_profits, linewidth=1, color="steelblue", alpha=0.7)
    plt.stackplot(dates, below_zero_profits, linewidth=1, color="red", alpha=0.7)

    plt.xlabel('Date', fontsize=18)
    plt.ylabel('Profit', fontsize=18)
    plt.title(f'Profit Over Time', fontsize=18)
    plt.grid(linestyle='--', color='grey', alpha=0.7)
    plt.axhline(0, linestyle="--", color="red")
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.show()
    plt.close()
```

## SMA CUMMULATIVE RETURN

```python
df['Daily Return 50'] = df['50-Day Moving Average'].pct_change()
df['Daily Return 200'] = df['200-Day Moving Average'].pct_change()
```

```python
df['Cumulative Return 50'] = (1 + df['Daily Return 50']).cumprod()
df['Cumulative Return 200'] = (1 + df['Daily Return 200']).cumprod()
```

```python
df[:10]
```

```python
initial_investment = 100000
```

## CUMMULATIVE RETURN SMA 50 & 200

```python
current_value = initial_investment * df['Cumulative Return 50'][-1]
if current_value > initial_investment:
    print(f"Profit: {initial_investment - current_value}")
    print(f"Current value: {current_value}")
else:
    print(f"Loss: {current_value - initial_investment}")
    print(f"Current value: {current_value}")
```

```python
current_value = initial_investment * df['Cumulative Return 200'][-1]
if current_value > initial_investment:
    print(f"Profit: {initial_investment - current_value}")
    print(f"Current value: {current_value}")
else:
    print(f"Loss: {current_value - initial_investment}")
    print(f"Current value: {current_value}")
```

```python
plt.figure(figsize=(15, 6))
df['Cumulative Return 50'].plot()
df['Cumulative Return 200'].plot()
plt.xlabel('Date', fontsize=16)
plt.ylabel('Cumulative Return', fontsize=16)
plt.grid(linestyle="--", color="grey")
plt.ticklabel_format(style='plain', axis='y')
plt.legend()
plt.show()
```
